# Univariate Analysis

```python
# Iterate over the columns for plotting
for i, predictor in enumerate(df[[
    "BRId", "GenderId", "IAId", "Nationality", "Occupation", "Fee Structure", 
    "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"
]].columns):
    
    # Create a figure with better visibility
    plt.figure(figsize=(10, 6))

    # Set the title dynamically based on the column name
    plt.title(f"Distribution of {predictor}", fontsize=16, pad=20)

    # Add labels for axes
    plt.xlabel(predictor, fontsize=12)
    plt.ylabel("Count", fontsize=12)

    # Rotate x-axis labels for readability
    plt.xticks(rotation=45, ha="right", fontsize=10)

    # Create the countplot
    ax = sns.countplot(data=df, x=predictor, width=0.75,legend=False)
    #sns.color_palette("Set2")
    # Add value labels on top of the bars
    for p in ax.patches:
        ax.annotate(
            f'{int(p.get_height())}', 
            (p.get_x() + p.get_width() / 2., p.get_height()),
            ha='center', va='center', fontsize=8, color='black', 
            xytext=(0, 10), textcoords='offset points'
        )

    # Add horizontal gridlines for better readability
    plt.grid(True, axis="y", linestyle="--", alpha=0.7)

    # Display the plot
    plt.show()
```

# Numerical Analysis

```python

# Select the relevant columns and remove missing values
columns = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans',
           'Bank Deposits','Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending' ]

# Plot histograms using a for loop
plt.figure(figsize=(20,20))

for i, col in enumerate(columns):
    ax=plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True,ax=ax)
    #plt.hist(df[col], kde=True)
    #plt.xlabel(col)
    #plt.xlabel(col)
   # plt.ylabel("Frequency")
    
    plt.grid(True, axis="y", linestyle="--", alpha=0.7)
    plt.title(col)
    for container in ax.containers:
        ax.bar_label(container, fmt='%d', fontsize=6, color='black', padding=3)
plt.tight_layout()
plt.show()
```

```python
#Another way of creating Histogram
# Select relevant columns 
columns = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans',
           'Bank Deposits','Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending' ]

# Plot histograms using a for loop
plt.figure(figsize=(10, 20))  # Adjust figure size

for i, col in enumerate(columns):
    ax=plt.subplot(len(columns), 1, i + 1)
    # Uncomment below if using seaborn for kde
    # sns.histplot(df[col], kde=True)
    bars=plt.hist(df[col], bins=20, color="green", edgecolor="black", alpha=0.7)
    plt.xlabel(col)
    #plt.ylabel("Frequency")
    plt.title(f"Histogram of {col}")
    plt.grid(True, axis="y", linestyle="--", alpha=0.7)
     # Add value labels
    for bar in bars[2]:  # bars[2] contains the patches (bars)
        height = bar.get_height()
        if height > 0:  # Only annotate bars that are visible
            ax.annotate(f'{int(height)}',
                        xy=(bar.get_x() + bar.get_width() / 2., height),
                        xytext=(0, 5),  # Offset labels slightly above bars
                        textcoords="offset points",
                        ha='center', va='center',
                        fontsize=8, color='black')

# Show all plots together
plt.tight_layout()  # Adjust spacing between plots
plt.show()
```

# HeatMap by numerical cols


```python

# Select relevant columns and remove missing values
columns = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans',
           'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

# Compute correlation matrix
corr_matrix = df[columns].corr()

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Title and display
plt.title("Heatmap of Numerical columns")
plt.show()
```

# **insights from  correlation heatmap:**

#### **1. Strong Positive Correlations**
- **Bank Deposits & Checking Accounts (0.84)**: This indicates a strong relationship between deposits and checking accounts, suggesting that individuals who maintain higher deposits likely have active checking accounts.
- **Bank Deposits & Saving Accounts (0.75)**: People who save regularly also tend to maintain substantial deposits, showing a habit of structured financial planning.
- **Business Lending & Bank Deposits (0.44)**: Businesses that secure loans may also maintain higher deposit balances, possibly due to strategic financial management.

#### **2. Moderate Positive Correlations**
- **Estimated Income & Superannuation Savings (0.37)**: A moderate correlation suggests that higher-income individuals tend to allocate funds toward superannuation savings, but it's not a direct one-to-one relationship.
- **Credit Card Balance & Bank Deposits (0.38)**: Interestingly, individuals with high credit card balances may also maintain larger deposits, possibly due to a combination of spending behavior and structured financial reserves.
- **Bank Loans & Business Lending (0.42)**: Businesses seeking loans may rely on a mix of personal and company financial resources, which results in a moderate correlation between individual bank loans and business lending.

#### **3. Weak or No Significant Correlation**
- **Foreign Currency Account & Checking Accounts**: If this correlation is weak, it suggests that people who hold foreign currency accounts may not necessarily be frequent users of checking accounts in their domestic banking system.
- **Credit Card Balance & Superannuation Savings**: If these two show little correlation, it indicates that credit card spending habits are independent of retirement saving strategies.

#### **4. Potentially Negative or Reverse Trends**
- If **Bank Loans & Saving Accounts** show a weak or negative correlation, it could indicate that individuals with high loan burdens are less likely to maintain substantial savings.
- **Credit Card Balance & Checking Accounts** may show lower correlation, suggesting that daily transaction accounts are not necessarily linked to higher credit card usage.

#### **What This Means for Financial Decision-Making**:
- Strong savings habits appear interconnected (Checking Accounts, Deposits, and Savings).
- Business loans and individual banking behaviors have some degree of correlation.
- People may not always save in superannuation accounts in direct proportion to their income.
